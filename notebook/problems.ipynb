# Probability distribution problems

## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Write the above functions but for both normal and uniform distributions.

Distribucion normal

```python
from scipy.stats import norm

def dnorm(x, mean=0, std=1):
    result = norm.pdf(x, loc=mean, scale=std)
    return result

def pnorm(q, mean=0, std=1):
    result = norm.cdf(q, loc=mean, scale=std)
    return result

def qnorm(p, mean=0, std=1):
    result = norm.ppf(p, loc=mean, scale=std)
    return result

def rnorm(n, mean=0, std=1):
    result = norm.rvs(loc=mean, scale=std, size=n)
    return result
```

```python
np.random.seed(42)

mean = 0
std = 1

print(f"Densidad de probabilidad de una Normal(0, 1) en x = 2: {dnorm(2, mean=mean, std=std)}")
print(f"Probabilidad acumulada de una Normal(0, 1) para x < 2: {pnorm(2, mean=mean, std=std)}")
print(f"Valor de una Normal(0, 1) con una probabilidad acumulada de 0.9: {qnorm(0.9, mean=mean, std=std)}")
print(f"Generar 2000 valores aleatorios de una Normal(0, 1): {rnorm(2000, mean=mean, std=std)}")
```

```python
np.random.seed(42)

data = np.linspace(-5, 5, 1000) 
probabilities = np.linspace(0, 1, 1000)

pdf = norm.pdf(data, loc=mean, scale=std)
cdf = norm.cdf(data, loc=mean, scale=std)
ppf = norm.ppf(probabilities, loc=mean, scale=std)

fig, axes = plt.subplots(3, 1, figsize=(5, 9))

axes[0].plot(data, pdf, "r-")
axes[0].set_title('Función de densidad de probabilidad (Normal)')
axes[0].set_ylabel('Densidad')

axes[1].plot(data, cdf, "b-")
axes[1].set_title('Función de distribución acumulativa (Normal)')
axes[1].set_ylabel('Probabilidad acumulada')

axes[2].plot(probabilities, ppf, "g-")
axes[2].set_title('Función de punto porcentual (Normal)')
axes[2].set_ylabel('Valor')
axes[2].set_xlabel('Probabilidad')

plt.tight_layout()

plt.show()
```

Distribucion uniforme

```python
from scipy.stats import uniform

def dunif(x, low=0, high=1):
    result = uniform.pdf(x, loc=low, scale=high - low)
    return result

def punif(q, low=0, high=1):
    result = uniform.cdf(q, loc=low, scale=high - low)
    return result

def qunif(p, low=0, high=1):
    result = uniform.ppf(p, loc=low, scale=high - low)
    return result

def runif(n, low=0, high=1):
    result = uniform.rvs(loc=low, scale=high - low, size=n)
    return result
```

```python
np.random.seed(42)

low = 0
high = 1

print(f"Densidad de probabilidad de una Uniforme(0, 1) en x = 0.5: {dunif(0.5, low=low, high=high)}")
print(f"Probabilidad acumulada de una Uniforme(0, 1) para x < 0.5: {punif(0.5, low=low, high=high)}")
print(f"Valor de una Uniforme(0, 1) con una probabilidad acumulada de 0.9: {qunif(0.9, low=low, high=high)}")
print(f"Generar 2000 valores aleatorios de una Uniforme(0, 1): {runif(2000, low=low, high=high)}")
```

```python
from scipy.stats import uniform

def calcular_pdf(data, low, high):
    return uniform.pdf(data, loc=low, scale=high - low)

def calcular_cdf(data, low, high):
    return uniform.cdf(data, loc=low, scale=high - low)

def calcular_ppf(probabilities, low, high):
    return uniform.ppf(probabilities, loc=low, scale=high - low)


data = np.arange(-5, 5, 0.01)
probabilities = np.linspace(0, 1, len(data))

pdf = calcular_pdf(data, low, high)
cdf = calcular_cdf(data, low, high)
ppf = calcular_ppf(probabilities, low, high)

fig, axes = plt.subplots(3, 1, figsize=(5, 7))

axes[0].plot(data, pdf, "r-")
axes[0].set_title('Función de densidad de probabilidad (Uniforme)')
axes[0].set_ylabel('Densidad')

axes[1].plot(data, cdf, "b-")
axes[1].set_title('Función de distribución acumulativa (Uniforme)')
axes[1].set_ylabel('Probabilidad acumulada')

axes[2].plot(probabilities, ppf, "g-")
axes[2].set_title('Función de punto porcentual (Uniforme)')
axes[2].set_ylabel('Valor')
axes[2].set_xlabel('Probabilidad')


plt.tight_layout()

plt.show()
```
